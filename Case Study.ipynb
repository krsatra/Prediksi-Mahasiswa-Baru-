```python
import pandas as pd
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_absolute_error

# Membaca data dari file CSV
data = pd.read_csv('Data.csv')

# Memisahkan variabel independen (X) dan variabel dependen (y)
X_train = data[['Tahun', 'Uang_Pangkal', 'Akreditasi']].values
y_train = data['Jumlah_Mahasiswa'].values

# Membangun model SVR
regressor = SVR(kernel='rbf')
regressor.fit(X_train, y_train)

# Data testing
tahun_2023 = 2023  # Tahun 2023
uang_pangkal_2023 = 1.1 * data['Uang_Pangkal'].iloc[1]  # Uang pangkal tahun 2023 ditambah 10% dari data 2022
akreditasi_2023 = 5  # Anggap akreditasi tetap 5

# Melakukan prediksi jumlah mahasiswa di tahun 2023
X_test = [[tahun_2023, uang_pangkal_2023, akreditasi_2023]]
y_pred = regressor.predict(X_test)

print("Prediksi Jumlah Mahasiswa di Tahun 2023:", y_pred[0])

# Data sebenarnya untuk menghitung akurasi
X_actual = data[['Tahun', 'Uang_Pangkal', 'Akreditasi']].values
y_actual = data['Jumlah_Mahasiswa'].values

# Melakukan prediksi pada data training
y_train_pred = regressor.predict(X_train)

# Menghitung koefisien determinasi (R^2) pada data training
r2_train = r2_score(y_train, y_train_pred)

# Melakukan prediksi pada data sebenarnya
y_actual_pred = regressor.predict(X_actual)

# Menghitung koefisien determinasi (R^2) pada data sebenarnya
r2_actual = r2_score(y_actual, y_actual_pred)

# Menghitung tingkat kesalahan rata-rata (MAE) pada data sebenarnya
mae = mean_absolute_error(y_actual, y_actual_pred)

print("Koefisien Determinasi (R^2) pada Data Training:", r2_train)
print("Koefisien Determinasi (R^2) pada Data Sebenarnya:", r2_actual)
print("Tingkat Kesalahan Rata-Rata (MAE) pada Data Sebenarnya:", mae)
```
